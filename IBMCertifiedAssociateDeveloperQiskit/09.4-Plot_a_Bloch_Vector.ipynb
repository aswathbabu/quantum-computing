## Plot a bloch vector


### Using cartesian coordinates

```python
from qiskit.visualization import plot_bloch_vector

coord = [1,0,0]

plot_bloch_vector(coord, figsize=(3,3), coord_type='cartesian')
```

```python
from qiskit.visualization import plot_bloch_vector

coord = [0, 1, 0]

plot_bloch_vector(coord, figsize=(3,3), coord_type='cartesian')
```

### Using spherical coordinates

```python
import numpy as np

coord = [1, np.pi/2, 0]

plot_bloch_vector(coord, figsize=(3,3), coord_type="spherical")
```

```python
import numpy as np

coord = [1, np.pi/2, np.pi/6]

plot_bloch_vector(coord, figsize=(3,3), coord_type="spherical")
```
