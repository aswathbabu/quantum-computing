## Plot error map


### For a backend

```python
from qiskit_ibm_provider import IBMProvider
from qiskit.visualization import plot_error_map

provider = IBMProvider(instance="ibm-q/open/main")
backend1 = provider.get_backend("ibmq_quito")
backend2 = provider.get_backend("ibmq_manila")

display(plot_error_map(backend1))
display(plot_error_map(backend2))
#Compare the errors on these two QCs. They have didferent processors
```
